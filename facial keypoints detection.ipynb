```python
from tensorflow.keras import Sequential,regularizers
from tensorflow.keras.layers import Dense
```

```python
def largemodelr():
    #forward pass
    model=Sequential()
    model.add(Dense(units=512,input_shape=[1],activation='relu',kernel_regularizer=regularizers.l2(l=0.001)))
    #here i have included activation
    model.add(Dense(units=1024,activation='relu',kernel_regularizer=regularizers.l2(l=0.001)))
    model.add(Dense(units=1024,activation='relu',kernel_regularizer=regularizers.l2(l=0.001)))
    model.add(Dense(units=512,activation='relu',kernel_regularizer=regularizers.l2(l=0.001)))   
    model.add(Dense(units=1,kernel_regularizer=regularizers.l2(l=0.001)))
    #backward pass
    model.compile(optimizer='adam',loss='mean_squared_error')
    # here the optimizer has become adam here the number of epochs is reduced
    # at starting it takes higher learning rate at the end the learning rate is reduced
    return model
```
